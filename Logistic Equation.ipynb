```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# logistic function 

def logistic (a, x0, N):    
    x = x0
    x_list = [x0]
    for i in range(N-1):      
        x = a * x * (1. - x)
        x_list.append(x)
    
    return x_list
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(16, 6), facecolor='lightgray')
plt.xlabel('No. of iterations')
plt.ylabel('Value')
#plt.title(f'\nLogistic Equation\n\na={a}  |  x0={x0}\n')
```

# 2. Steady State Behavior

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(16, 6), facecolor='lightgray')
plt.xlabel('No. of iterations')
plt.ylabel('Value')

plt.plot(logistic(2.5,0.5,10), 'o:r', marker='.')
plt.plot(logistic(2.5,0.2,10), 'o:r', color='g', marker='.')
plt.plot(logistic(2.5,0.9,10), 'o:r', color = 'b', marker='.')
```

# 3. Period Doubling

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(16, 6), facecolor='lightgray')
plt.xlabel('No. of iterations')
plt.ylabel('Value')

plt.plot(logistic(2.99,0.667,200), 'o:r',marker='.')
plt.plot(logistic(3.01,0.5,200), 'o:r', color='b',marker='.')
```
