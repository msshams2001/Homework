```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
def an_sw(n):
    return (-4/(np.pi*n))*(-1)**((n+1)/2)
```

```python
sw = np.zeros(50)
t = np.linspace(-0.5,0.5)
```

```python
# 1+3 term
for k in range(2):
    sw = sw + an_sw(2*k+1)*np.cos((2*k+1)*2*np.pi*t)
plt.plot(t, sw)
```
