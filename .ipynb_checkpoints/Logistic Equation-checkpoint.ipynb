```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# logistic function 

def logistic (a, x0, N):    
    x = x0
    x_list = [x0]
    for i in range(N-1):      
        x = a * x * (1. - x)
        x_list.append(x)
        
    plt.style.use('seaborn-whitegrid')
    plt.figure(figsize=(16, 6), facecolor='lightgray')
    plt.xlabel('No. of iterations')
    plt.ylabel('Value')
    plt.title(f'\nLogistic Equation\n\na={a}  |  x0={x0}\n')
    
    return x_list
```

# 1.

```python
logistic(2,2,20)
```

# 2. 

```python
logistic()
```
