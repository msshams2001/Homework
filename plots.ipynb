```python
import pandas as pd
df = pd.read_csv('part d.csv')
df.head(1)

from matplotlib import pyplot as plt
import numpy as np
import scipy.stats as stats 
import math
```

```python
data = df['Latest: Radiation (Counts)']
bins = np.arange(min(data),50,1)

plt.hist(df['Latest: Radiation (Counts)'],bins)
plt.title("Histogram of part 1 section D")
plt.xlabel("No. of Decay Counts")
plt.ylabel("Counts")

mu = 32.47
variance = 4.735
sigma = math.sqrt(variance)
x = np.linspace(0, 50, 1)
plt.plot(x, stats.norm.pdf(x, mu, sigma))

plt.show()
```
